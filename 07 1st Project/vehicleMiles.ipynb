```python
#Import Dependencies
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read file into dataframe
file = os.path.join("..","..","table_01_35q119_1.xlsx")
df = pd.read_excel(file)
df.head()
```

```python
#Get first row of data, totals
dfTotal = df.iloc[0,:]
#cut data to capture 2000-2016, in line with pollution data
dfTotal = dfTotal[1:len(dfTotal)-1]
#create dataframe
dfTotal=pd.DataFrame({"totalMiles":dfTotal})
dfTotal.head()
```

```python
#plot data as line
dfTotal.plot(kind="line")
plt.ylabel("Miles Driven (millions)")
plt.xlabel("Year")
plt.title("US Vehicle-Miles (Millions) - Highway")
plt.savefig("US Vehicle Miles.png")
```
